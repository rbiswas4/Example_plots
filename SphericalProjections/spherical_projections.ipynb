## Coordinates (ra, dec)

These must be in radians. The range for ra is [-np.pi, np.pi]
The range for dec is [-np.pi/2, np.pi/2]

```python
# Generate a bunch of random coordinates

num_samps = 1000

# We could generate in radians, but these shows what we have to do if we have coordinates in degrees
x = np.random.uniform(size=num_samps)*360.
x[x > 180] -= 360 # Do the wrap
y = np.random.uniform(low=-90., high=90, size=num_samps)
x = np.radians(x)
y = np.radians(y)
y /= np.cos(y)

# Let us also generate 2 other random properties
z = np.random.uniform(-1, 1, num_samps)
h = np.random.uniform(0, 1, num_samps)
```

## min(x), max(x)

```python
max(y), max(y)
```

```python
fig = plt.figure()
#ax = fig.a(111, projection='mollweide')
#ax.plot(x, y)
```

```python
ax = fig.add_subplot(111, projection='mollweide')
ax.plot(x, y, '.') # plot is always faster than scatter, but scatter has more features
```

### Like in a general scatter plot, make the sizes of points according to a variable h and the color with z

```python
from matplotlib import cm
```

```python
fig2 = plt.figure(figsize=(6,6))
ax2 = fig2.add_subplot(111, projection='mollweide')
#fig2, ax2 = plt.subplots(figsize=(6,6))
#ax2.set_ylim(-10, 10)
cc = ax2.scatter(x, y, c=z, cmap=cm.viridis, s=h)
fig2.colorbar(cc)
_ = ax2.set_yticklabels([])
_ = ax2.set_xticklabels([])
```
