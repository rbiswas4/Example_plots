```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_style('whitegrid')
sns.set_context('paper',  font_scale=2.5)
```

```python
data = np.random.chisquare(50, size=500000)
data2 = np.random.chisquare(60, size=500000)
```

```python
data
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
_ = ax.hist(data, bins=np.arange(0, 200, 1), histtype='step', label='data')
_ = ax.hist(data2, bins=np.arange(0, 200, 1), histtype='step', label='data2')
ax.legend(loc='upper right')
ax.set_xlabel('X')
ax.set_ylabel('count')
fig.tight_layout()
fig.savefig('myfig.pdf')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
_ = ax.hist(data, bins=np.arange(0, 200, 1), histtype='stepfilled', label='data')
_ = ax.hist(data2, bins=np.arange(0, 200, 1), histtype='stepfilled', label='data2', alpha=0.3)
ax.legend(loc='upper right')
ax.set_xlabel('X')
ax.set_ylabel('count')
fig.tight_layout()
fig.savefig('myfig.pdf')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
#_ = ax.hist(data, bins=np.arange(0, 200, 1), histtype='stepfilled', label='data')
#_ = ax.hist(data2, bins=np.arange(0, 200, 1), histtype='stepfilled', label='data2', alpha=0.3)
sns.kdeplot(data, ax=ax, label='data', shade=True)
sns.kdeplot(data2, ax=ax, label='data2', shade=True)
ax.legend(loc='upper right')
ax.set_xlabel('X')
ax.set_ylabel('count')
fig.tight_layout()
fig.savefig('myfig.pdf')
```
